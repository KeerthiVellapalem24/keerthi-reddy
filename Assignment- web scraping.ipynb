```python
import requests
from bs4 import BeautifulSoup

class Scrapper:
    
    def __init__(self,url,item):
        self.final_url= url + '/search?q='+item
        self.pull_data(self.final_url)
    
    def pull_data(self,finalurl):
        print(finalurl)
        self.result=requests.get(finalurl)
        #print('pull data function')
        
        self.soup=BeautifulSoup(self.result.content)
        #print(self.soup.prettify())
        self.tagline=self.soup.find('div',attrs={'class','_4rR01T'})
        print(self.tagline.text)
        
        self.price=self.soup.find('div',attrs={'class','_30jeq3 _1_WHN1'})
        print(self.price.text)
        
        self.ratings=self.soup.find('div',attrs={'class','_3LWZlK'})
        print(self.ratings.text)
        
        self.specifications=self.soup.find('div',attrs={'class','fMghEO'})
        print(self.specifications.text)
        
        self.exchange_offer=self.soup.find('div',attrs={'class','_2ZdXDB'}) 
        print(self.exchange_offer.text)
        
        self.alltext=self.soup.findAll('div',attrs={'class','_3pLy-c row'})
        count=0
        for item in self.alltext:
            count=count+1
            print(item.text)
            print('***'*50)
        else:
            print('total items in page={}'.format(count))
        
        
    def refine_data(self):
        pass
                            
```
